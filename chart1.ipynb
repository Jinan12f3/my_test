```python
path = r"C:\Users\1\Desktop\scanning.xlsx"
data = pd.read_excel(path)
data.head()
```

```python
time_dict = {
    "Monday": 0,
    "Tuesday": 1,
    "Wednesday": 2,
    "Thursday": 3,
    "Friday": 4,
    "Saturday": 5,
    "Sunday": 6,
}
data_list = list()
ts = list(data[u"扫描时间戳"])
for t in ts:
    h = str(time.strftime("%H", time.localtime(int(str(t)))))
    a = str(time_dict.get(str(time.strftime("%A", time.localtime(int(str(t)))))))
    data_list.append(a+h)

fun = lambda x: "0%s" % str(x) if int(x) < 10 else str(x)
datastr3 = ""
for i in range(7):
    for j in range(24):
        ls = [i, j]
        s = str(i)+fun(j)
        print s
        ls.append(data_list.count(s))
        datastr3 += str(ls) + ","
datastr3 = datastr3[:-1]
```
