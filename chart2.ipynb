```python
path = r"C:\Users\1\Desktop\scanning.xlsx"
data = pd.read_excel(path)
data.head()
```

```python
newD = list(data[u"扫描机型"])
data_dict = dict()
for i in newD:
    if i not in data_dict.keys():
        data_dict[i] = 1
    else:
        data_dict[i] += 1
```

```python
datastr1 = ""
datastr2 = ""

for k in data_dict.keys():
    s1 = "\'%s\'" % k
    datastr1 += s1 + ","
    
datastr1 = datastr1[:-1]

for d in data_dict:
    s2 = "{value: %s, name: \'%s\'}" % (data_dict[d], d)
    datastr2 += s2 + ","
    
datastr2 = datastr2[:-1]
```
